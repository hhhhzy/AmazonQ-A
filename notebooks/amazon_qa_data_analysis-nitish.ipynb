```python
import nltk

def n_grams(x, n):
    d = {}
    for sentence in x:
        n_grams = nltk.ngrams(sentence.lower().replace(',', '').split(), n)
        for i in n_grams:
            d[i] = d.get(i, 0) + 1
    return pd.DataFrame.from_dict(d, orient='index').sort_values(0, ascending=False)
```

```python
bigrams = n_grams(reviews_table.head(100000).reviewText.values, 2).iloc[:1000]
trigrams = n_grams(reviews_table.head(100000).reviewText.values, 3)[:1000]
```
